```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Salaries.csv')

#  Salary Distribution by Rank and Discipline (Box Plot)
plt.figure(figsize=(12, 6))
sns.boxplot(x='rank', y='salary', hue='discipline', data=df)
plt.title('Salary Distribution by Rank and Discipline')
plt.show()

```

```python


#  Salary vs. Years of Service (Scatter Plot with Regression Line)
plt.figure(figsize=(10, 6))
sns.regplot(x='service', y='salary', data=df)
plt.title('Salary vs. Years of Service')
plt.show()
```

```python
#  Distribution of Years Since PhD by Rank (Violin Plot)
plt.figure(figsize=(10, 6))
sns.violinplot(x='rank', y='phd', data=df)
plt.title('Distribution of Years Since PhD by Rank')
plt.show()
```

```python

#  Gender Distribution within Each Discipline (Stacked Bar Plot)
gender_distribution = df.groupby(['discipline', 'sex']).size().unstack().fillna(0)
gender_distribution.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Gender Distribution within Each Discipline')
plt.xlabel('Discipline')
plt.ylabel('Count')
plt.show()
```

```python

#  Salary by Years Since PhD and Gender (Line Plot)
plt.figure(figsize=(12, 6))
sns.lineplot(x='phd', y='salary', hue='sex', data=df)
plt.title('Salary by Years Since PhD and Gender')
plt.show()
```

```python
#  Correlation Matrix (Heatmap)
correlation_matrix = df[['salary', 'phd', 'service']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
#  Salary Distribution by Rank and Gender (Box Plot)
plt.figure(figsize=(12, 6))
sns.boxplot(x='rank', y='salary', hue='sex', data=df)
plt.title('Salary Distribution by Rank and Gender')
plt.show()
```

```python
#  Salary by Discipline and Years of Service (Facet Grid with Scatter Plots)
g = sns.FacetGrid(df, col='discipline', col_wrap=2, height=5)
g.map(sns.scatterplot, 'service', 'salary')
g.set_axis_labels('Years of Service', 'Salary')
plt.show()
```
